```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/network_benchmarks.csv')
print('Dataset shape:', df.shape)
print(df.head(10))
```

```python
signal_cols = ['latency_score','bandwidth_score','ip_reputation_score','asn_trust_score','routing_quality_score','connection_stability_score']
signal_avgs = df[signal_cols].mean()
colors = ['#06b6d4','#3b82f6','#10b981','#f59e0b','#8b5cf6','#ef4444']
plt.figure(figsize=(12,5))
plt.bar(signal_avgs.index, signal_avgs.values, color=colors)
plt.title('SpeedCheck - Average Network Signal Scores', fontsize=14, fontweight='bold')
plt.xlabel('Network Signal')
plt.ylabel('Average Score (0-100)')
plt.xticks(rotation=20)
plt.axhline(y=60, color='green', linestyle='--', alpha=0.5, label='Healthy threshold')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
conn_counts = df['connection_type'].value_counts()
plt.figure(figsize=(8,5))
plt.bar(conn_counts.index, conn_counts.values, color=colors[:len(conn_counts)])
plt.title('SpeedCheck - Connection Type Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
